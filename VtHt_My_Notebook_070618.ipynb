`Ctrl Enter` exécute la cellule.  
`Shift Enter` exécute la cellule et passe à la cellule suivante en mode commande.  
*`Alt Enter` execute la cellule et passe à la cellule suivante en mode édition.

Lorsque l'on fait un retour chariot, python execute 3 opérations : 
>Premiere opération : il crée l'entier 3 dans l'eo

>Le type de la variable est lié à l'objet, python est un  langage à **typage fort dynamique**.Le type n'est pas lié à la variable qui référence l'objet mais à l'objet.

### CODAGE_JEUX_DE_CARACTERES_ET_UNICODE

11000011101100110100111000111100101  est un flux de bits  
>En informatique on va décoder ce flux de bits ici découpage en blocs de 7 bits en suivant le codage ASCII     
`97 108 105 99 101`  est la conversion décimale de ce flux de bits de gauche à droite  après décodage 7 bits     
`a  1   i   c  e`     jeux de caractere ASCII    
Ensuite pour etre capable d'afficher un caractère, il faut utiliser une police de caractères  
`a  l   i   c  e`     police lucida calligraphy  
Cette police de caractère va définir un glyphe, un dessin que l'on peut afficher à l'écran.  
On a fait une opération de décodage.
Malheureusement, il y a des jeux de caractères incompatibles. C'est pourquoi on a créé **UNICODE** pour coder l'intégralité des caractères du monde entier.  
En pratique, on utilise le codage **UTF8**. python est totalement compatible avec le codage unicode. mais il faut contrôler son encodage.  
**La règle : utiliser tout le temps utf8**  
**UNICODE** affiche l'intégralité des caratères du monde mais il faut une police qui supporte ce caractère, ie, qui est capable de dessiner un glyphe du caractère correspondant.

* ### Le Context Manager, instruction `with`

>Il est fondamental de fermer les fichiers ou alors il faut utiliser le **Context Manager** avec l'instruction `with` et dans ce cas on ne s'occupe pas de la fermeture des fichiers.

* ###### Créer et Ecrire dans un fichier avec le Context Mananger

````python
with open (r'/Users/vtht/essais/spam_cm.txt','w',encoding='utf8') as f_cm:
    for i in range(100):
        f_cm.write(f"ligne {i}\n")
````

````python
!cat /Users/vtht/essais/spam_cm.txt
````

* ###### Lire dans un fichier avec le Context Mananger

````python
with open(r'/Users/vtht/essais/spam.txt','r',encoding='utf8') as f_cm:
    print(f_cm.read())
````

````python
with open(r'/Users/vtht/essais/spam_cm.txt','r',encoding='utf8') as f_cm:
    with open(r'/Users/vtht/essais/spam_cm_mod.txt','w',encoding='utf8') as f_cm_mod:
        for line in f_cm:
            line=line.split()
            line[0]=line[0].upper()
            f_cm_mod.write(",".join(line)+"\n")
````

* ### Packing et unpacking

>***Le packing et l'unpacking sont deux concepts qui offrent la possibilité de trnasformer une succession d'arguments, nommés ou non, en liste/tuple ou dictionnaire et vice-versa.***

* #### ***Unpacking et Packing implicites***

>***Unpacking implicite*** fonctionne avec des tuples, listes, strings, et tout autre iterable.

````python
#Le tuple(2,3) va être depaqueté et sauvegardé dans les variables a et b.
a,b=(2,3) #Tuple unpacking.
````

>***Packing implicite*** fonctionne avec des tuples, listes, strings, et tout autre iterable.

````python
#Les valeurs de a et b sont empaquetées dans une tuple t.
t=(a,b) #Tuple packing
type(t)
````

>L'opérateur splat `*` n'a pas été utilisé ici, il s'agit d'Unpacking et de Packing implicites.

* #### ***Unpacking et Packing explicites***

>***Unpacking explicite avec une liste et un dictionnaire***

````python
def aire_rectangle(cote1=0,cote2=0):
    return cote1*cote2
````

````python
rec1=[3,8] 
print (aire_rectangle(*rec1)) #La liste rec1 est dépaquetée en arguments unitaires
rec2={'cote1':4,'cote2':8}
print (aire_rectangle(**rec2)) #Le dictionnaire rec2 est dépaqueté en arguments unitaires
````

>***Packing explicite avec une liste et un dictionnaire***

````python
def aire_rectangle1(*args): #Les arguments passés en paramètres sont paquetés dans args qui se comporte comme un tuple 
    if len(args)==2:
        return args[0]*args[1]
    else:
        print('Merci de stipuler deux paramètres')
````

````python
#Une liste va être créée à partir des arguments fournis
print (aire_rectangle1(3,8))
````

````python
def aire_rectangle2(**kwargs): #Les arguments passés en paramètres sont paquetés dans args qui se comporte comme un dictionnaire
    if len(kwargs)==2:
        result=1
        for key,value in kwargs.items():
            result*=value
        return result
    else:
        print('Merci de stipuler deux paramètres')
````

````python
#Un dictionnaire va être crée à partir des arguments nommés
print(aire_rectangle2(cote1=4,cote2=8))
````

* La portée d'une variable détermine de quel endroit du code on peut accéder à cette variable.

* ***la portée lexicale des variables*** est la façon dont python trouve les variables, ie, en fonction de **l'endroit dans le code*** où cette variable a été définie.

* ***Une variable locale est une variable définie dans le bloc d'instruction d'une fonction.***   
*** Une variable locale est détruite avec un return de la fonction ou à l'appel de la fonction.***

* ***Une variable globale est une variable définie en dehors de toute fonction.***

* ***La règle `LEGB:Localement Englobante Globalement Buitins`***
>* `Localement` : On regarde si la variable a été définie là où elle est référencée, sinon :
>* `Englobante` : On remonte de la fonction la plus proche de là où on référence la variable jusqu'à la fonction la plus éloignée, sinon :
>* `Globalement` : on regarde au niveau des variables globales, sinon :
>* `Builtins` : on regarde dans le module `Builtins`.

* ### Exemple

````python
a,b,c=1,1,1
def g():
    b,c=2,4
    b=b+10
    def h():
        c=5
        print(a,b,c)
    h() #Appel de la fonction h
g() #Appel de la fonction g
````

>* **Les variables** `a,b,c`  
>`a` est une variable globale, `a`vaut `1`  
>`b`est définie localement dans la fonction englobante `g`, b vaut `12` 
>`c`est définie localement dans la fonction `h`, `c`vaut `5`    
>* **Rappel sur les variables**   
>Une variable définie dans une fonction devient locale à cette fonction. Elle ne peut pas être vue en dehors de cette fonction.  
>Une variable globale définie au niveau du module peut être vue par toutes les fonctions définies dans ce module.

````python
print(a,b,c)
````

>En effet les variables `a,b,c`sont définies globalement et le `print`est appelé au niveau du module.

* #### Création d'un itérateur

````python
class Phrase:
    def __init__(self, ma_phrase):
        self.ma_phrase=ma_phrase
        self.mots=ma_phrase.split()
    def __iter__(self):
        return self
    def __next__(self):
        if not self.mots:
            raise StopIteration #permet de lever l'exception StopIteration
        return self.mots.pop(0)
````

````python
class Phrase:
    def __init__(self,ma_phrase):
        self.ma_phrase=ma_phrase
        self.mots=ma_phrase.split()
    def __iter__(self):
        return IterPhrase(self.mots)
    
class IterPhrase:
    def __init__(self,mots):
        self.mots=mots[:]
    def __iter__(self):
        return self
    def __next__(self):
        if not self.mots:
            raise StopIteration
        return self.mots.pop(0)
````

* **Exemple** : On  va créer la `class Timer`qui permet d'avoir la durée d'execution du bloc de code et on va utiliser la class avec le context manager appelé avec l'instruction `with`

````python
import time
class Timer:
    def __enter__(self):
        self.start=time.time()
        return self
    def __exit__(self,*args):
        duree=time.time()-self.start
        print(f"{duree}s")
        return False
    def __str__(self):
        duree=time.time()-self.start
        return f"intermediaire: {duree}s"
````

* ## `np.linspace, np.trapz`

`np.linspace(a,b,N)` : permet de distribuer `N` points régulièrement espacés sur un intervalle `[a,b]`. Si l'argument `retstep=True` est donné, alors renvoie la largeur des intervalles en plus du tableau de points.  
`np.trapz(fx,x)` : permet de faire l'intégration par la méthode des trapèzes à partir d'une série `f(x)` des valeurs d'une fonction prise en des points `x`. Le tableau `x` est optionnel, on peut aussi définir la valeur `dx=h` si les points sont espacés d'une largeur `h` fixée.

````python
def integration_trapz(f,a,b,n=100):
    '''Retour sur l'intégration par méthode des trapèzes en délégant tout le travail à np.trapz(). Rappel : n est le nombre d'intervalles et non le nombre de points(cf pb clôtures/piquets)'''
    # On génère d'abord les n+1 points=régullièrement espacés sur [a,b]
    x_arr=np.linspace(a,b,n+1)
    # Puis les valeurs de la fonction à intégrer sur ces points x
    fx_arr=[f(x) for x in x_arr] # trapz() fera la traduction "au vol" en array
    #Le premier argument est le tableau des valeurs de la fonction à intégrer
    #Le deuxième argument est les positions où ont été prises ces valeurs
    return np.trapz(fx_arr,x_arr) 
    
    
````

* ### Résolution d'équation différentielle

On utilise comme méthode  `odeint(f,y0,t)` pour intégrer numériquement une équation différentielle du type $\dfrac{dy}{dt}=f(y,t)$
* **La fonction inconnue `y`** est
>soit un système d'équations couplées
 soit une équation d'ordre supérieur vue comme un système d'équations couplées
* **`y0`** est
>la liste des conditions initiales pour chaque élément du tableau de valeurs y à l'instant t(0).  
* **`t`** représente
>l'ensemble des temps dans lequel on veut résoudre l'équation différentielle. 


On traite les 3 exemples suivants, 4) et 5) sont introduits pour matplotlib ultérieurement :  
* 1)$\dfrac{dx}{dt}=-\dfrac{x}{\tau}$ 
* 2)$\dfrac{dx_1}{dt}=-k\cdot x_2$ et $\dfrac{dx_2}{dt}=-k\cdot x_1$
* 3)$\ddot{x}+2\lambda\dot{x}+\omega_0^2 x=0$

>il s'écrit$\dfrac{dx}{dt}=\dot x$  et $\dfrac{d\dot x}{dt}=-2\lambda\dot x-\omega_0^2 x$
* 4)$\ddot{x}+2\lambda\dot{x}+\omega_0^2 x = A\cos\omega t$ 
>il s'écrit $\dfrac{dx}{dt}=\dot x$ et $\dfrac{d\dot x}{dt}=-2\lambda\dot x-\omega_0^2 x+A\cos\omega t$ 
* 5)$\dddot{x}+\alpha\ddot{x}+2\lambda\dot{x}+\omega^2_0 x=0$
>il s'écrit $\dfrac{dx}{dt}=\dot x$ et $\dfrac{d\dot x}{dt}=\ddot x$ et $\dfrac{d\ddot x}{dt}=-\alpha\ddot x-2\lambda\dot x-\omega_0^2 x$

````python
#ETUDE DE 1)
import scipy as sp
import scipy.integrate
#y est x
#La solution du 1) doit donner expt(-t/tau)
def f(y,t):
    tau=1.0
    return -y/tau
y0=1 #Condition initiale x(0)=1
t=range(4) #Points temporels pour les calculs
sp.integrate.odeint(f,y0,t) 
# Intégration de f(y,t), y0 condition initiale, t variable d'integration
````

* ### Analyse de Fourier

On traite le signal suivant : 
* **Signal** $\mathbf{s(t)=10\cos (2\pi f_1t)+5\cos (2\pi f_2t)}$ avec $\mathbf{f_1=2Hz}$ et $\mathbf{f_2=8Hz}$.  

>On échantillonne ce signal sur l'intervalle $t\in [0s,12s]$ en y répartissant de manière régulièrement espacée 600 points.

* **Echantillonage** $\mathbf{t\in [0s,12s],600pts}$ soit `t=sp.linspace(0,12,600)` 

>On peut décider d'ajouter un bruit aléatoire en prenant un `sp.random.random` de `len(t)` points, ici 600 points dans l'intervalle `[0.0,1.0]` auxquels on retranche 0.5 pour avoir des valeurs réparties dan l'intervalle `[-0.5,0.5]`. Pour avoir un bruit significatif on multiplie les amplitudes de ces points par 50 pour avoir 600 points répartis sur `[-25.,25.]`.  
On crée ce bruit avec 

* **Bruit**  `bruit=50*(sp.random.random(len(t))-0.5)` 

>On applique l'algorithme de transformée de Fourier rapide FFT avec `sp.fft()` qui renvoie des valeurs complexes dont on ne s'intéresse qu'au module d'où l'usage de `abs()` selon

* **FFT** `FFT_module=abs(sp.fft())`

>On récupère les fréquences d'échantillonnage avec la fonction `sp.fftpack.fftfreq()` dont la deuxième moitié des valeurs est négative. C'est pourquoi on ne prend que la moitié des valeurs pour l'affichage. 

* **Fréquences** `freqs=sp.fftfreq(signal.size,t[1]-t[0])`


````python
%matplotlib inline


import scipy as sp #La base
import scipy.fftpack #Pour récupérer les fréquences de la FFT
import matplotlib.pyplot as plt #Pour faire les schémas
#Définition des temps pour l'échantillonnage
t=sp.linspace(0,12,600)
#Le signal pur
signal=10*sp.cos(2*sp.pi*2.0*t)+5*sp.cos(2*sp.pi*8.0*t)
bruit=50*(sp.random.random(len(t))-0.5)
signal_avec_bruit=signal+bruit
#On fait les FFT sur le signal et sur le signal_avec_bruit en prenant le module.
#Pour avoir les amplitudes, on normalise en divisant par la moitié du nombre de points disponibles
FFT_signal=abs(sp.fft(signal))/(signal.size/2)
FFT_signal_avec_bruit=abs(sp.fft(signal_avec_bruit))/(signal_avec_bruit.size/2)
freqs=sp.fftpack.fftfreq(signal.size,t[1]-t[0])
#On ne prend pas en compte les résultats pour les fréquences négatives
FFT_signal=FFT_signal[0:len(FFT_signal)//2]
FFT_signal_avec_bruit=FFT_signal_avec_bruit[0:len(FFT_signal_avec_bruit)//2]
freqs=freqs[0:len(freqs)//2]
#On fait les schémas avec matplotlib
def fais_graphique(signal,FFT,freqs,titre):
    with plt.xkcd(): #XkcD sketch style, for the fun of it!
        plt.figure(dpi=150)
        plt.subplot(211) #Fenêtre en 2 lignes, 1 colonne, première figure 
        plt.title(titre)
        plt.plot(t,signal) #Le signal en fonction du temps
        plt.xlabel('Temps $t$ en s')
        plt.ylabel('Signal en V')
        plt.subplot(212) #Fenêtre en 2 lignes, 1 colonne, deuxième figure
        plt.plot(freqs,FFT) #La FFT en fonction de la fréquence
        plt.xlabel('Fréquence $f$ en Hz')
        plt.ylabel('FFT en V')
        plt.tight_layout() #Pour mettre de l'espace entre les sous-figures
        #plt.savefig(fichier) Sauvegarde
        #plt.clf() #Nettoyage
        #plt.show()
fais_graphique(signal,FFT_signal,freqs,"FFT sur un signal pur")
fais_graphique(signal_avec_bruit,FFT_signal_avec_bruit,freqs,"FFT sur un signal bruité")

````

## Recherche des racines d'une fonction

Le module `scipy.optimize` contient des procédures pour trouver des minima, des zeros ou des points fixes.  
Voici deux algorithmes de recherche de racines : 
`sp.optimize.newton(f,x0,fprime=None)`, fprime=dérivée de $f$, soit $f'$
>Soit une fonction $f$ et $x_0$ le voisinage d'une de ses racines, cette méthode essaie de calculer la position du zero, soit   
>* par la méthode de la sécante(cas ou $f'$ n'est pas donnée)  
>* par la méthode de Newton(cas où $f'$ est donnée)

>Avantage : On n'a pas besoin d'encadrer la racine dans un intervalle $[a;b]$ tel que $f(a)f(b)<0$.

## "Fittage" d'un modèle sur des données expérimentales

`sp.optimize.curve_fit(f,x,data,p0,sigma)`  
>Cette méthode essaie de faire coller la fonction $f$ de variable principale $x$ (avec comme variables secondaires les paramètres d'ajustement souhaités) sur les données mesurées $data$.  
Les arguments $p0$ et $sigma$sont optionnels.  
>* `p0` représente le jeu de paramètres initiaux servant de points de départ à la procédure d'optimisation (par défaut, $p0=1$).  
>* `sigma` doit être un tableau de même taille que $data$ et est utilisé comme poids statistique dans la régression(cas où certaines mesures sont moins précises que d'autres).  

>Cette méthode renvoie un doublet(tableau 1D,tableau 2D).   
>* tableau 1D représente les valeurs optimales des paramètres d'ajustement par la méthode des moindres carrés.  
>* tableau 2D représente la matrice de covariance dont les éléments sur la diagonale donnent la variance du paramètre d'ajustement correspondant.

````python
import numpy as np
import scipy as sp
import scipy.optimize
#Définition de notre fonction ax+b
def fonction_lineaire(x,a,b):
    return a*x+b
#Distribution des points en x
x=np.linspace(0,10,100)
#Les "mesures" parfaites
y=fonction_lineaire(x,3,18)
#On rajoute du bruit
data=y+0.2*np.random.normal(size=len(x))
#L'ajustement et l'affichage des paramètres ajustés et des covariances
p,pcov=sp.optimize.curve_fit(fonction_lineaire,x,data)
p
````

````python
#On veut retrouver la courbe de distribution des notes.

import numpy as np
import scipy as sp
import scipy.optimize
import matplotlib as mpl
import matplotlib.pyplot as plt
import random # Tirage de notes au hasard
#Gaussienne de moyenne 10 et d'écart-type 3 pour toutes les pcsi
notes=[random.gauss(10,3) for k in range(130)]
#On demande à numpy de faire l'histogramme "à la main".
hist_data,bin_edges=np.histogram(notes,bins=20,range=(0,20))

#Edge représente le bord
#On prend la hauteur milieu pour la valeur des notes
bin_centers=(bin_edges[:-1]+bin_edges[1:])/2
#On définit le modèle à fitter ici une gaussienne
def gauss(x,A,mu,sigma):
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
#On définit les valeurs initiales des paramètres du modèle
A0_mu_0_sigma_0=[30,8,6]
coeff,var_matrix=sp.optimize.curve_fit(gauss,bin_centers,hist_data,p0=A0_mu_0_sigma_0)
A,mu,sigma=coeff #Récupération des paramètres
hist_fit=gauss(bin_centers,A,mu,sigma) #La courbe fittée obtenue
rmu,rsigma=round(mu,2),round(sigma,2) #coeff avec deux chiffres décimaux
with plt.xkcd(): #XkcD sketch style, for the fun of it!
        plt.figure(dpi=150) #Resolution de l'image
        plt.hist(notes,bins=20,range=(0,20),label='Notes')
        plt.plot(bin_centers,hist_fit,label='Gaussienne, mu={}, sigma={}'.format(rmu,rsigma))
        plt.legend() #Choix automatique du placement de la légende
        plt.ylim(0,30) #Pour laisser de la place pour la légende
        plt.show
        #plt.savefig('fit_result.png') #Sauvegarde
        #plt.clf() #Nettoyage
````

* ### ***Barres d'erreurs***

En TIPE, il est utile de présenter des résultats signalants les barres d'erreurs associées aux mesures expérimentales.  

`plt.errorbar(x,y,xerr,yerr)`
>Cette méthode permet de tracer en plus du graphe les barres d'erreurs.  
Les barres d'erreur peuvent être
>* Soit de taille identique fixée par un nombre
>* Soit de taille individuelle, selon un axe, fixée par un tableau 1D de taille `len(x)`
>* Soit de taille individuelle, selon les deux axes, fixée par un tableau 2D de taille `2*len(x)`

````python
# On reprend l'exemple de fittage affine
import numpy as np
import scipy as sp
import scipy.optimize
import matplotlib as mpl
import matplotlib.pyplot as plt
f=lambda x,a,b:a*x+b #La fonction
x=np.linspace(0,10,6) #La distribution des points en x
y=f(x,3,18) #Les mesures parfaites
datay=y+2*np.random.normal(size=len(x)) #On ajoute du "bruit" en y
datax=x+0.2*np.random.normal(size=len(x)) #On ajoute du "bruit" en x
#Ajustement avec les mesures parfaites x, bruité datay
p1,p1cov=sp.optimize.curve_fit(f,x,datay)
#Rappel=la forme * unpacke les arguments de p1
fit1=f(x,*p1) #On échantillonne la fonction "fittée"
#Ajustement avec les mesures bruitées datax, bruitées datay
p2,p2cov=sp.optimize.curve_fit(f,datax,datay)
#Rappel=la forme * unpacke les arguments de p1
fit2=f(datax,*p2) #On échantillonne la fonction "fittée"
#On représente le tout avec des barres d'erreur à 2 sigma
a1,b1,a2,b2=[round(a,2) for a in (p1[0],p1[1],p2[0],p2[1])] #deux décimales
with plt.xkcd(): #XkcD sketch style, for the fun of it!
    plt.figure(dpi=150) #Resolution de l'image
    plt.subplot(211) #Fenêtre en 2 lignes, 1 colonne, première figure
    plt.errorbar(x,datay,yerr=4,label='y avec du bruit',fmt='x') #fmt='x' est le choix du marker pour le point
    plt.plot(x,fit1,label='fit linéaire : a={}, b={}'.format(a1,b1))
    plt.xlim((-2,12)) #On change les limites en x pour voir les bords
    plt.legend(loc='upper left',fontsize='xx-small') #Légende en haut à gauche
    plt.subplot(212) #Fenêtre en 2 lignes, 1 colonne, deuxième figure
    plt.errorbar(datax,datay,xerr=0.4,yerr=4,label='x,y avec du bruit',fmt='o') #fmt='x' est le choix du marker pour le point
    plt.plot(datax,fit2,label='fit linéaire : a={}, b={}'.format(a2,b2))
    plt.xlim((-2,12)) #On change les limites en x pour voir les bords
    plt.legend(loc='upper left',fontsize='xx-small') #Légende en haut à gauche
    plt.show()
````

* ### ***Exemples de résolution d'équations différentielles***

* 4)$\ddot{x}+2\lambda\dot{x}+\omega_0^2 x = A\cos\omega t$ 
>il s'écrit $\dfrac{dx}{dt}=\dot x$ et $\dfrac{d\dot x}{dt}=-2\lambda\dot x-\omega_0^2 x+A\cos\omega t$ 

````python
# Résolution du système 4) et affichage
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.integrate

def f(y,t):
    x,xpoint=y
    lamb, w0,A,w=0.01,1,2,3
    return [xpoint, -2*lamb*xpoint-w0**2*x+A*np.cos(w*t)]
t=sp.linspace(0,30,1001) #1001 points de mesures
y0=[1,0] # x(0)=1 et dx/dt(0)=0
sol=sp.integrate.odeint(f,y0,t) #Resolution
position=sol[:,0] #Chalocopie de la liste de liste, premier argument de chaque liste
vitesse=sol[:,1] #Chalocopie de la liste de liste, deuxième argument de chaque liste
with plt.xkcd(): #XkcD sketch style, for the fun of it!
    plt.figure(dpi=150) #Resolution de l'image
    plt.subplot(211) #Fenêtre en 2 lignes, 1 colonne, première figure
    plt.plot(t,position,label='Position (en m)')
    plt.plot(t,vitesse,label='Vitesse (en m/s)')
    plt.legend(loc='upper left',fontsize='xx-small')
    plt.xlabel('Temps (en s)')
    plt.subplot(212) #Fenêtre en 2 lignes, 1 colonne, deuxième figure
    plt.plot(position,vitesse)
    plt.xlabel('Position (en m)')
    plt.ylabel('Vitesse (en m/s)')
    plt.show()

````

````python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as ipyw
def sinus(freq,amp):
    X=np.linspace(0.,4*np.pi,200)
    Y=amp*np.sin(freq*X)
    with plt.xkcd(): #XkcD sketch style, for the fun of it!
        plt.figure(dpi=150) #Resolution de l'image
        plt.xlim((0,12)) #On change les limites en x pour voir les bords
        plt.ylim((-6,6)) #On change les limites en x pour voir les bords
        plt.plot(X,Y)
        
ipyw.interact (sinus,freq=(0.2,10.,0.1),amp=(1.,5.,0.2))
plt.show()
````

>* `value=1` est la position initiale de la réglette

````python
# les widgets pour construire le tableau de bord
from ipywidgets import (FloatSlider,interactive_output,IntSlider, Dropdown, Layout,HBox, VBox, Text)
from IPython.display import display
def sinus4(freq, phase, amplitude, domain):
    """"
    une fonction qui visualise sinus avec 4 réglages
    la fréquence, la phase, l'amplitude,
    et le domaine sur lequel on fait le dessin
    (en nombre de pis)
    """

    X = np.linspace(0., domain*np.pi, 500)
    Y = amplitude * np.sin(freq*(X+phase))

    # il faut fixer la taille de l'axe en Y
    # pour que les animations soient plus parlantes
    # sinon, quand on change l'amplitude, il ne passe rien
    # graphiquement car alors la taille de l'axe est calculée
    # automatiquement
    with plt.xkcd(): #XkcD sketch style, for the fun of it!
        plt.figure(dpi=150) #Resolution de l'image
        plt.axis([0, domain*np.pi, -12, 12])
        plt.plot(X, Y)
    # cette fonction un peu absconse permet de
# créer le tableau de bord en assemblant
# et en arrangeant des widgets
def dashboard():
    """
    create and display a dashboard
    return a tuple with
    the dashboard itself, and
    a dictionary name->widget suitable for interactive_output
    """
    # dashboard pieces as widgets
    l_75 = Layout(width='75%')
    l_50 = Layout(width='50%')
    l_25 = Layout(width='25%')

    w_freq = Dropdown(options=list(range(1, 10)),
                      value = 1,
                      description = "fréquence",
                      layout=l_50)
    w_phase = FloatSlider(min=0., max = 2*np.pi, step=np.pi/12,
                          description="phase",
                          value=0., layout=l_75)
    w_amplitude = Dropdown(options={"micro" : .1,
                                    "mini" : .5,
                                    "normal" : 1.,
                                    "grand" : 3.,
                                    "énorme" : 10.},
                           value = 3.,
                           description = "amplitude",
                           layout = l_25)
    w_domain = IntSlider(min=1, max=10, description="dom. n * π", layout=l_50)

    # make up a dashboard
    dashboard = VBox([
                      HBox([w_domain, w_freq]),
                      HBox([w_amplitude, w_phase]),
                     ])
    return dashboard, dict(freq=w_freq, phase=w_phase, amplitude=w_amplitude, domain=w_domain)
# interactively call sinus4
ui, env = dashboard()
out = interactive_output(sinus4, env)
display(ui, out)
# en l'état il reste un petit bug
# il faut bouger au moins un des réglages
# pour que ça commence à s'afficher
````
