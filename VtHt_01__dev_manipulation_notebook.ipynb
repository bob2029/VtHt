# 1.  MANIPULATION DES NOTEBOOKS

Le notebook est constitué d'une succession de cellules. Les cellules peuvent être dans le mode commande(liseré bleu) ou en mode edition(liseré vert).  
* ## LE MODE EDITION (LISERE VERT)  
>Pour entrer dans le mode édition de la cellule sélectionnée, il suffit de presser la touche `Enter` ou de cliquer à l'intérieur de la cellule. Quand une cellule est en mode édition, on saisit du texte comme dans un éditeur classique.  
Lorsque le curseur est en début de ligne, la touche `tab` indente et la touche `Shift-Tab` désindente.  
* #### RACCOURCIS CLAVIER EN MODE EDITION  
>`Ctrl A` ou `Cmd A` sélectionne tout le texte de la cellule.  
`Ctrl Z` ou `Cmd Z` annule les dernières saisies du texte.  
`Enter` passe dans le mode édition.  
`ESC` passe dans le mode commande.  
`Ctrl Enter` exécute la cellule.  
`Shift Enter` exécute la cellule et passe à la cellule suivante en mode commande.  
*`Alt Enter` execute la cellule et passe à la cellule suivante en mode édition.  

* ## LE MODE COMMANDE (LISERE BLEU)  
>En mode commande, on peut ajouter ou supprimer de cellules mais on ne peut pas saisir de texte dans la cellule. Attention dans ce mode commande, certaines touches du clavier sot associées à des actions sur le notebook. Tenter de saisir du texte dans ce mode peut donc produire des effets inattendus.  

* #### RACCOURCIS CLAVIER EN MODE COMMANDE  
>`a` insère une nouvelle cellule au-dessus de la cellule sélectionnée.  
`b` insère une nouvelle cellule au-dessous de la cellule sélectionnée.  
`m` transforme la cellule en une cellule de type Markdown.  
`y` transforme la cellule en une cellule de type Code.  
`c` copie des cellules.  
`v` colle des cellules.  
`dd` supprime la cellule sélectionnée.  
`z` annule la dernière suppression de cellule.  
`ii` interrompt l'execution du code.  
`00`  (zerozero)redémarre l'interpréteur. Il se retrouve alors dans son état initial.  
`h` affiche la liste de tous les raccourcis clavier.  
Si on a modifié des raccourcis et que l'on souhaite faire un reset il suffit d'effacer le contenu du fichier ~/.jupyter/nbconfig/notebook.json  
* ## LE LANGAGE MARKDOWN   
>le langage Markdown est un langage de balisage léger qui permet de rédiger du texte enrichi. Il est possible aussi d'inserer du LateX.  

* ## L_INTERPRETEUR PYTHON  
>Lorsque l'on ouvre un notebook, on le visualise tel qu'il a été sauvegardé, penser du coup à nettoyer le kernel avec 00 pour avoir un fichier plus léger. A l'ouverture, un nouvel interpréteur est lancé.  
* #### LES EXCEPTIONS  
>L'interpreteur peut s'interrompre en levant une exception. Pour comprendre l'exception, il faut identifier son nom(ex IndexError), lire le message associé (ex 'list index of range') et identifier l'expression qui l'a provoqué. On peut aussi avoir une inteeruption de code (KeyboardInterrupt) si le code ne se termine pas (calcul infini).  
* #### EXECUTION D_UNE CELLULE    
>lorsqu'une cellule de code s'execute le prompt comporte une étoile In(etoile).  
* #### LE_TERMINAL_VIA_IPYTHON  
>Si le texte de la cellule démarre par un `!`, ce n'est pas du code python mais une commande bash.  
* #### LES_COMMANDES_MAGIQUES_DE_IPYTHON  

`%pylab inline`  
>Importe les modules numpy et matplotlib. L'option inline indique que les figures Matplotlib seront insérées dans le notebook lui-même. Cette instruction execute en particulier deux instructions :  
import numpy as np  
import matplotlib.pyplot as plt  

`%%writefile`   
>Crée un fichier de texte. voici un exemple :   
%%writefile square.py  
k=1  
s=1  
...  
Overwriting square.py  

`%less`  
>Cette commande permet de visualiser le contenu d'un fichier de texte.  
%less square.py  

`%run`  
>Cette commande execute le script comme dans un terminal selon :  
%run square py  
equivalent à  
python3 square.py dans un terminal.  
La différence est qu'à la fin de l'execution du script tous les objets définis dans le script sont insérés dans l'espace de nommage global de l'interpréteur.  

`%timeit`  
>Mesure le temps d'execution d'une instruction. Le code est executé plusieurs fois et le temps moyen d'execution est affiché. % timeit function  

`%pastebin`  
>Envoie une texte vers le serveur Github Gist. Tout le monde peut alors lire le texte à partir de l'url retournée par la commande.
In() %pastebin arithmetique.py  
Out() 'https://gist.github.com/0409dc7e9be57c539ffbf481a67c1c61'  

`%who et %whos`  
>Affiche les objets qui sont actuellement définis dans l'interpréteur python.  
In() %who  
T X euclide hauteur etc  
In() %whos  
Les mets sous forme d'un tableau (Variable,Type,Data/Info)

* ## EXPORTER_UN_NOTEBOOK
>Enregistrer en local avec File/Download as...  
Publier dans la partie publique de votre dossier personnel(~/pub/).  
exemple, le notebook se trouve dans~/nb.ipynb, on ouvre un terminal et on execute :   
jupyter nbconvert ~/nb  
cp ~/nb ~/pub  
La première commande a créer un fichier nb.html dans le dossier où se trouve le notebook. La seconde commande a copié le fichier html dans le dossier/pub/. Quiconque peut consulter votre fichier désormais à l'adresse https://moncompte/nb.html.  Pour produire un fichier pdf, il suffit de faire  
jupyter nbconvert --to pdf ~/nb.
